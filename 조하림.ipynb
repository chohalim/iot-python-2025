#### 1번

```python
def knapsack():
    array = [[0 for _ in range(maxWeight+1)] for _ in range(rowCount+1)]
    for row in range(1,rowCount+1):
        for col in range(1,maxWeight+1): 
            if weight[row]>col: 
                array[row][col] = array[row-1][col] 
            else:
                value1 = money[row] + array[row-1][col-weight[row]]
                value2 = array[row-1][col]
                array[row][col] = max(value1, value2)

    return array[rowCount][maxWeight]

maxWeight = 7
rowCount = 4
weight = [0, 6, 4, 3, 5]
money = [0, 13, 8, 6, 12]

maxValue = knapsack()
print()
print('배낭에 담을 수 있는 보석의 최대 가격 ->', maxValue, '억원')
```

```python
# 더이상 삽입 못할때
def isQueueFull(): # 큐가 다 찾을 때(사이즈만큼)
    global SIZE, queue, front, rear 
    while (rear == SIZE-1): 
        return True
    else:
        return False

# 더이상 추출 못할때
def isQueueEmpty(): # 큐가 비었을 때
    global SIZE, queue, front, rear
    while front == rear:
        return True
    else:
        return False

# 더이상 삽입 못할때
def enQueue(data):
    global SIZE, queue, front, rear
    while isQueueFull():
        print('Queue is full!')
    else:
        rear += 1
        queue[rear] = data

# 더이상 추출 못할때
def deQueue():
    global SIZE, queue, front, rear 
    while isQueueEmpty():
        print('Queue is empty')
        return None
    else:
        front += 1
        data = queue[front]
        queue[front] = None
        for i in range(front+1, SIZE):
            queue[i-1] = queue[i] # front 추출하고 나서 한칸씩 앞으로 보내짐
            queue[i] = None

        front -= 1
        rear -= 1
        return data
    
def peek():
    global SIZE, queue, front, rear
    while isQueueEmpty():
        print('Queue is empty')
        break
    else:
        return queue[front+1]
    
# 초기화
SIZE = int(input('큐크기 입력 > ')) # 입력한 값 SIZE로 받음
queue = [None for _ in range(SIZE)] # 큐를 SIZE만큼 다 None로 채움
front = rear = -1 # 앞 뒤는 -1로 시작

## 메인 모듈
if __name__ == '__main__':
    while True:
        select = input('삽입(I)/추출(E)/확인(V)/종료(Q) --> ').upper()

        if select == 'Q': 
            break
        elif select == 'I': 
            data = input('데이터 입력 > ')
            enQueue(data)
            print(f'큐 상태 : {queue}')
        elif select == 'E':
            data = deQueue()
            print(f'추출 데이터 : {data}')
            print(f'큐 상태 : {queue}')
        elif select == 'V':
            data = peek()
            print(f'확인 데이터 : {data}')
            print(f'큐 상태 : {queue}')
        else:
            print('입력 오류')
```

```python
import random
memory = []
root = None
dataAry = ['콘말차', '삿포로 맥주', '아카페라 벤티 헤이즐넛', '레어치즈푸딩',
            '척오리지널 메가 사워', '요아정 요거트컵',
              '페퍼로니피자주먹밥', '널담 슈톨렌', 
              '딸기마시멜로케이크', '버니공쥬 딸기뚱카롱', '고추잡채호빵', '체다 슈크림붕어스낵']
sellAry = [random.choice(dataAry) for _ in range(20)]
print('오늘 판매된 물건(중복o)->', sellAry)

# TreeNode클래스 선언
class TreeNode:
    def __init__(self):
        self.left = None
        self.data = None
        self.right = None

# 메인 모듈
if __name__ == '__main__':
    node = TreeNode()
    node.data = dataAry[0]
    root = node
    memory.append(node)

    for name in dataAry[1:]:
        node = TreeNode()
        node.data = name

        current = root
        while True:
            if name < current.data: 
                if current.left == None:
                    current.left = node
                    break 
                else:
                    current = current.left 
            else: 
                if current.right == None:
                    current.right = node
                    break
                else:
                    current = current.right

        memory.append(node)

    print('이진탐색 트리 구성완료!') 

    findName = input('찾을 이름 입력 > ')

    current = root
    count = 0
    while True:
        if findName == current.data:
            print(f'{findName} 찾음!')
            count += 1
            
            break
        elif findName < current.data:
            if current.left == None:
                print(f'{findName} 트리에 없음')
                break
            else:
                current = current.left
                count += 1
        else:
            if current.right == None:
                print(f'{findName} 트리에 없음')
                break
            else:
                current = current.right
                count += 1

    print(f'검색 종료 : 총 검색횟수 {count}')
    
                

def preorder(node):
    if node == None:
        return
    print(node.data,end=' ')
    preorder(node.left)
    preorder(node.right)

print('오늘 판매된 종류(중복x)->',end=' ')
preorder(root)
```

```python
def binSearch(ary, fData):
    pos = -1
    start = 0
    end = len(ary)-1
    count = 0

    while (start <= end):
        mid = (start+end)//2 # 반으로 나눠서 찾는다
        count += 1 
        if fData == ary[mid]: # 찾는 수를 찾았을 때
            print(f'총 검색횟수 {count}')
            return mid 
        elif fData > ary[mid]:
            start = mid+1
        else:
            end=mid-1

    print(f'총 검색횟수 {count}') # 찾는 수가 dataAry 뽑히지 않았을 때
    return pos
    

import random
Ary = [i for i in range(1,100001)] # 범위는 1~100000까지의 수를
dataAry = [random.choice(Ary) for _ in range(100000)] # 100000번 랜덤으로 수를 뽑는다
dataAry.sort() # 뽑은 수를 오름차순 정렬
findData=int(input('찾는 데이터를 입력하세요 ->')) 

position = binSearch(dataAry,findData) 
if position == -1: 
    print(f'{findData}(이) 없네요.')
else:
    print(f'{findData}는 {position}위치에 있음')
```
